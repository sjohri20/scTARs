```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Human Brain 2

```python
%time
b2_genes_celltags=pd.read_csv("/home/dbeb/btech/bb1160039/scratch/project/scanpy results/Human Brain 2/genes_celltags.csv", index_col=0)
b2_ngenes_celltags=pd.read_csv("/home/dbeb/btech/bb1160039/scratch/project/scanpy results/Human Brain 2/non_gene_celltags.csv", index_col=0)
```

```python
b2_pivot_frame=pd.DataFrame({"gene_clusters":b2_genes_celltags.leiden,"ngene_clusters":b2_ngenes_celltags.leiden}).dropna()
```

```python
sns.heatmap(pd.pivot_table(b2_pivot_frame, columns=["gene_clusters"],index=["ngene_clusters"], aggfunc=len, fill_value=0),annot=True, fmt="d",cmap="YlGnBu")
```
